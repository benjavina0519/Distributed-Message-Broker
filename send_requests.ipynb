```python
import pandas as pd
import requests
import json

# Define API endpoint
API_URL = "http://localhost:8000/submit"

# Read CSV file
csv_file_path = "Python Coding Exercise - Input Data.csv"  # Change to your actual CSV file path
df = pd.read_csv(csv_file_path)

# Loop through each row and send a POST request
for index, row in df.iterrows():
    raw_text = row["raw_text"]  # Ensure this matches the CSV column name
    
    # Prepare JSON payload
    payload = {"raw_text": raw_text}
    
    # Send POST request
    response = requests.post(API_URL, json=payload)
    
    # Print response
    print(f"Sent: {payload} | Response: {response.status_code}, {response.text}")
```
